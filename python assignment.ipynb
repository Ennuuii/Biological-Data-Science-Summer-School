```python
import matplotlib.pyplot as plt
with open(r'C:\Users\user\Downloads\new 24.txt', 'r') as file:
    s = file.read().replace(',', '.').replace('	', ' ').replace('\n', ' ')
    k = s.split( )
    iht = list(map(float, k))
    odd_i = []
    even_i = []
    for i in range(0, len(iht)):
        if i % 2:
            even_i.append(iht[i])
        else:
            odd_i.append(iht[i])
    average = sum(even_i) / len(even_i)
    d = average
    e = []
    for h in (even_i):
        if h >= d:
            pass
        else: 
            e.append(h)
p = e[0]
m = even_i.index(p)
y = even_i[m: (-m)]
x = odd_i[m: (-m)]
plt.figure(figsize=(10,15))
plt.plot(x, y,label= ".",color='blue', marker='.', markersize = 7)
plt.xlabel('x ')
plt.ylabel('y ')
plt.title('графік залежності значень')
plt.savefig("графік залежності значень.png", facecolor='white', edgecolor='white') 
```
